# Transformata DCT a unei imagini

Transformata DCT se extinde la mai multe dimensiuni similar cu transformata DFT. Pentru un semnal bidimensional, precum o imagine, DCT-II devine:

$$
Y_{m_1,m_2} = \sum_{n_1=0}^{N_1-1}
              \sum_{n_2=0}^{N_2-1}
              x_{n_1,n_2}
              \cos\left[\frac{\pi}{N_1}m_1
                \left(n_1 + \frac12\right)\right]
              \cos\left[\frac{\pi}{N_2}m_2\left(n_2 + \frac12\right)\right]
$$

* unde $n_1$ și $n_2$ sunt pozițile pixelilor pe orizontală, respectiv, pe verticală
* bin-urile rezultate corespund pozițiilor pixelilor
* spectrul este în continuare simetric și par
* proprietățile transformatei DCT-II sunt respectate și în cazul celei 2D

În Python avem rutina `scipy.fft.dct` pentru 1D și `scipy.fft.dctn` pentru generalizarea la semnale $n$-dimensionale. Dimensiunea este determinată automat după forma semnalului; tipul DCT poate fi specificat prin atributul `type` (implicit II).

```python
Y1 = dctn(X, type=1)
Y2 = dctn(X, type=2)
Y3 = dctn(X, type=3)
Y4 = dctn(X, type=4)
freq_db_1 = 20*np.log10(abs(Y1))
freq_db_2 = 20*np.log10(abs(Y2))
freq_db_3 = 20*np.log10(abs(Y3))
freq_db_4 = 20*np.log10(abs(Y4))

plt.subplot(221).imshow(freq_db_1)
plt.subplot(222).imshow(freq_db_2)
plt.subplot(223).imshow(freq_db_3)
plt.subplot(224).imshow(freq_db_4)
plt.show()
```

# JPEG

Algoritmul de compresie JPEG are patru etape:

1. transformarea imaginii din pixeli RGB în Y'CbCr
2. aplicarea 2D-DCT pe blocuri disincte de 8x8 pixeli din imagine
3. cuantizarea în frecvență cu $Q$ dat de standardul JPEG
4. compresia rezultatului cu coduri Huffman

Unde matricea JPEG de cuantizare $Q$ este:
$$
Q =
\begin{bmatrix}
16 & 11 & 10 & 16 & 24 & 40 & 51 & 61 & \\
12 & 12 & 14 & 19 & 26 & 28 & 60 & 55 & \\
14 & 13 & 16 & 24 & 40 & 57 & 69 & 56 & \\
14 & 17 & 22 & 29 & 51 & 87 & 80 & 62 & \\
18 & 22 & 37 & 56 & 68 & 109 & 103 & 77 & \\
24 & 35 & 55 & 64 & 81 & 104 & 113 & 92 & \\
49 & 64 & 78 & 87 & 103 & 121 & 120 & 101\\
72 & 92 & 95 & 98 & 112 & 100 & 103 & 99\\
\end{bmatrix}
$$

Imaginea noastră de test este monocromă, deci nu necesită pasul 1, dar putem efectua o operație de *down-sampling* în preprocesare precum am prezentat la curs.

```python

Q_down = 10

X_jpeg = X.copy()
X_jpeg = Q_down*np.round(X_jpeg/Q_down);

plt.subplot(121).imshow(X, cmap=plt.cm.gray)
plt.title('Original')
plt.subplot(122).imshow(X_jpeg, cmap=plt.cm.gray)
plt.title('Down-sampled')
plt.show()
```

Pentru fiecare bloc de $8\times 8$ aplică DCT și cuantizare.

```python
Q_jpeg = np.array([[16, 11, 10, 16, 24, 40, 51, 61],
          [12, 12, 14, 19, 26, 28, 60, 55],
          [14, 13, 16, 24, 40, 57, 69, 56],
          [14, 17, 22, 29, 51, 87, 80, 62],
          [18, 22, 37, 56, 68, 109, 103, 77],
          [24, 35, 55, 64, 81, 104, 113, 92],
          [49, 64, 78, 87, 103, 121, 120, 101],
          [72, 92, 95, 98, 112, 100, 103, 99]])

# Encoding
x = X[:8, :8]
y = dctn(x)
y_jpeg = Q_jpeg*np.round(y/Q_jpeg)

# Decoding
x_jpeg = idctn(y_jpeg)

# Results
y_nnz = np.count_nonzero(y)
y_jpeg_nnz = np.count_nonzero(y_jpeg)

plt.subplot(121).imshow(x, cmap=plt.cm.gray)
plt.title('Original')
plt.subplot(122).imshow(x_jpeg, cmap=plt.cm.gray)
plt.title('JPEG')
plt.show()

print('Componente în frecvență:' + str(y_nnz) + 
      '\nComponente în frecvență după cuantizare: ' + str(y_jpeg_nnz))
```

```python
def show_and_save():
    global plt_index
    plt.savefig(f"plot_{plt_index}.pdf")
    plt_index += 1
    plt.show()


def jpeg(X, show=False, quant_factor=1):
    h, w = X.shape
    h_lim = (h // 8) * 8
    w_lim = (w // 8) * 8
    
    X_jpeg = np.copy(X)
    X_cut = X[:h_lim, :w_lim]

    # (h_lim, w_lim) -> (Rows//8, 8, Cols//8, 8) -> (Rows//8, Cols//8, 8, 8)
    blocks = X_cut.reshape(h_lim // 8, 8, w_lim // 8, 8).transpose(0, 2, 1, 3)

    Y = dctn(blocks, axes=(2, 3))
    
    orig_nnz = np.count_nonzero(Y)

    Q_j = Q_jpeg * quant_factor
    Y_jpeg = Q_j * np.round(Y / Q_j)
    
    jpeg_nnz = np.count_nonzero(Y_jpeg)

    X_rec_blocks = idctn(Y_jpeg, axes=(2, 3))

    # (Rows//8, Cols//8, 8, 8) -> (Rows//8, 8, Cols//8, 8) -> (h_lim, w_lim)
    X_rec_ordered = X_rec_blocks.transpose(0, 2, 1, 3).reshape(h_lim, w_lim)
    
    X_jpeg[:h_lim, :w_lim] = X_rec_ordered

    if show:
        plt.figure(figsize=(12, 6))
        plt.subplot(121).imshow(X, cmap=plt.cm.gray)
        plt.title('Original')
        plt.axis('off')
        
        plt.subplot(122).imshow(X_jpeg, cmap=plt.cm.gray)
        plt.title('Jpeg')
        plt.axis('off')
        
        plt.show()
        print(f"Original nonzero: {orig_nnz}\nJpeg nonzero: {jpeg_nnz}\n")

    return X_jpeg

_ = jpeg(X, show=True)
        
```

```python
def ycbcr2rgb(X_ycbcr):
    Y  = X_ycbcr[:, :, 0]
    Cb = X_ycbcr[:, :, 1] - 128
    Cr = X_ycbcr[:, :, 2] - 128

    R = Y + 1.402 * Cr
    G = Y - 0.344136 * Cb - 0.714136 * Cr
    B = Y + 1.772 * Cb

    # clamp values to 0-255
    R = np.clip(R, 0, 255)
    G = np.clip(G, 0, 255)
    B = np.clip(B, 0, 255)

    RGB = np.stack([R, G, B], axis=2).astype(np.uint8)
    return RGB
    
def rgb2ycbcr(X_rgb):
    R = X_rgb[:, :, 0]
    G = X_rgb[:, :, 1]
    B = X_rgb[:, :, 2]

    Y  =  0.299    * R + 0.587    * G + 0.114    * B
    Cb = -0.168736 * R - 0.331264 * G + 0.5      * B + 128
    Cr =  0.5      * R - 0.418688 * G - 0.081312 * B + 128

    YCbCr = np.stack([Y, Cb, Cr], axis=2)
    return YCbCr

def jpeg_rgb(X_rgb, show=False):
    X_ycbcr = rgb2ycbcr(X_rgb)
    X_jpeg = np.copy(X_ycbcr)
    X_jpeg[:, :, 0] = jpeg(X_jpeg[:, :, 0])
    X_jpeg[:, :, 1] = jpeg(X_jpeg[:, :, 1])
    X_jpeg[:, :, 2] = jpeg(X_jpeg[:, :, 2])
    X_jpeg = ycbcr2rgb(X_jpeg)

    if show:
        plt.subplot(121).imshow(X_rgb, cmap=plt.cm.gray)
        plt.title('Original')
        plt.subplot(122).imshow(X_jpeg)
        plt.title('Jpeg')
        plt.show()

    return X_jpeg
X_rgb = datasets.face()
_ = jpeg_rgb(X_rgb, show=True)
```

```python
def jpeg_mse(X, target=1):
    MSE = target + 1
    quant_factor = 1
    X_jpeg = None
    while MSE > target:
        X_jpeg = jpeg(X, quant_factor=quant_factor)
        MSE = np.mean((X_jpeg - X) ** 2)
        print(f"==============\nMSE: {MSE}\nquant_factor: {quant_factor}")
        quant_factor /= 2

    return X_jpeg
        
X_jpeg = jpeg_mse(X, target=.5)
plt.subplot(121).imshow(X, cmap=plt.cm.gray)
plt.title('Original')
plt.subplot(122).imshow(X_jpeg, cmap=plt.cm.gray)
plt.title('Jpeg')
plt.show()
    
```

```python
def jpeg_video(path):
    V = iio.imread(path)
    V_jpeg = []
    for frame in tqdm(V, "Frames processed:"):
        V_jpeg.append(jpeg_rgb(frame))
    
    return V_jpeg

V_jpeg = jpeg_video("video.mp4")
imageio.mimwrite('video_jpeg.mp4', V_jpeg, fps=30, codec='libx264')
        

```
